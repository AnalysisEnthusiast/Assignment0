```python
#----------------************Problem 1**************---------------------
import numpy as np 
import matplotlib.pyplot as plt 

#File Reading
with open('test.txt') as f:
    content = f.read()

#Converting Upper to Lower Cases & Sorting (Alphabetically)
contents=content.lower()
c=sorted_characters = sorted(contents)
# Filtering Non-Letters!
c = list(filter(('\n').__ne__, c))
c = list(filter((' ').__ne__, c))
c = list(filter((',').__ne__, c))
c = list(filter(('.').__ne__, c))

#Dictionary
counter = {}
let=[]
counti=[]
for letter in c:
    if letter not in counter:
      counter[letter] = 0
    counter[letter] += 1

print("Dictionary:\n",counter)
if len(counter)==26:
  print("\n****Great! All the Letters -A to Z- are available here!****\n") 
else:
  print('Oh, no! We do not have some of the letters!')

#List
for letter in counter:
  let.append(letter)
  counti.append(counter[letter])
print("Numpy Array:",let)
#print("Frequencies:"counti)

sum=0
perc=[]
for i in counti:
  sum=sum+i
for i in range (0,len(counti)):
  percentage=counti[i]/sum
  perc.append(percentage)
sumi=0
#print("Normalized Freguencies:",perc)
for i in perc:
  sumi=sumi+i

# Finding 3 highest values
from collections import Counter 
k = Counter(counter)
high = k.most_common(3)
print("\n Most Frequent Letters in the Provided Text File:\n Letters:Frequencies:\n")
for i in high:
    print(i[0]," :",i[1]," ")

# creating the bar plot

#Percentages 
fig = plt.figure(figsize = (10, 8))
X = let
Ys = perc
X_axis = np.arange(len(X)) 
plt.bar(X_axis - 0, Ys, 0.4,color='orange') 
plt.xticks(X_axis, X)
plt.xlabel("Letters")
plt.ylabel("Normalized Frequency")
plt.title("Appearance Rate of Different Letters Provided in the File")
plt.show()

#Numbers - Frequencies
fig = plt.figure(figsize = (10, 8))  
X = let
Ys = counti
X_axis = np.arange(len(X))
plt.bar(X_axis - 0, Ys, 0.4,color='m')
plt.xticks(X_axis, X)
plt.xlabel("Letters")
plt.ylabel("Frequency")
plt.title("Frequency Magnitude of Different Letters Provided in the File")
plt.show()
```

```python
#----------------************Problem 2**************---------------------
#************************Part 1
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
#Image Reading
img = cv2.imread('test.png')		# this is read in BGR format
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)		# this converts it into RGB
# matplotlib interprets images in RGB format, but OpenCV uses BGR format
print("********The Initial Image is:")
plt.imshow(rgb_img)
plt.show()

#Distance Computing
a=[0,0,250]
b=[0,250,0]
c=[250,0,0]
distl=[]
from scipy.spatial import distance
for indx , x in enumerate (img):
  for indy, y in enumerate (x):
    dista=(distance.euclidean(y, a))
    distb=(distance.euclidean(y, b))
    distc=(distance.euclidean(y, c))
    distg=min(dista,distb,distc)
#Replacing Part    
    if distg==dista:
      img[indx,indy]=a
    elif distg==distb:
      img[indx,indy]=b
    else:
      img[indx,indy]=c

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)		# this converts it into RGB
print("********The Converted Image is:")
plt.imshow(rgb_img)
plt.show()
cv2.imwrite('output1.png', img)
```
